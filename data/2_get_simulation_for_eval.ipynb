```python
from pyclmuapp import usp_clmu
import shutil
usp = usp_clmu() # will create a temporary working directory workdir

SURF="surfdata_1x1_JP-Yoy_detailed_simyr2000_c230710.nc"
FORCING="CTSM_DATM_JP-Yoy_2006-2020.nc"

shutil.copytree('fsrc', usp.input_path+'/usp/SourceMods/src.clm', dirs_exist_ok=True)
```

```python
usp_res = usp.run(
            case_name = "roofsp_raw",
            SURF = SURF,
            FORCING = FORCING,
            RUN_STARTDATE = "2006-01-01", # the start date of the simulation, must include in the forcing file time range
            STOP_OPTION = "nyears", # can be 'ndays', 'nmonths', 'nyears', 'nsteps'; nsteps means 1800s
            STOP_N = "15", # run for 10 years
            hist_type = "COLS", # output type
        )
```
